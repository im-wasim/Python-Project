<b>Word clouds (also known as text clouds or tag clouds) work in a simple way: 
the more a specific word appears in a source of textual data (such as a speech, blog post, or database),
the bigger and bolder it appears in the word cloud</b>

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

#Each pyplot function makes some change to a figure: e.g., creates a figure,
#creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. 
```

```python
movie =  pd.read_csv("movies_complete.csv")
```

###### Dropping NA values

```python
title_name =  movie['title'].dropna()
```

###### concatatting columns value into single value

```python
title_corpus =  " ".join(title_name)
title_corpus
```

###### genratting wordcloud

```python
title_wordcloud = WordCloud().generate(title_corpus)
#or
# title_wordcloud = WordCloud(background_color='white', height=200,width = 400 ,max_words=300).generate(title_corpus)
```

```python
title_wordcloud
```

###### Displaying wordcloud

```python
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud, interpolation= "kaiser") #used to display data as an image; i.e. on a 2D regular raster.
plt.axis('off')
plt.show()
```
